## Stochastic processes
### Brownian motion simulation

Brownian motion simulation where sampling for one particle motion is done using normal distributions N(0,dt), with mean 0 and variance described by dt (delta time).

This simple example illustrates how processes change over time, even though the first starting might be known.

NOTE for easier use this could be written into a class object.



```python
#Import libraries
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors
import random
```

```python
#Brownian motion simulation

def BrownianMotionSim(time,samples=100,seed=None):

    """
    Function takes time (s) as a range for simulation and performs brownian motion simulation 

    #Arguments 
        time, int/float: time in seconds
        samples, int: number of samples to generate brownian motion
        seed, int: you can set seed for reproducibility, default value is none
    #Returns
        array of distances (float) sampled from normal distribution N(0,dt)
    """
    #NOTE you may set seed for reproducibility
    if seed:
        set.seed(seed)
    t=time
    dt=t/samples

    distances=np.cumsum(np.random.normal(0,np.sqrt(dt),(1,samples)),axis=1)#we take samples for only one process where mean is 0 and variance is dt

    return distances
```

```python
def plotBrownian(time,n=5,samples=100,seed=None):

    """
    Function takes time (s) as a range for simulation and performs brownian motion simulation plotting

    #Arguments 
        time, int/float: time in seconds
        samples, int: number of samples to generate brownian motion
        seed, int: you can set seed for reproducibility, default value is none
        n, int: number of processes to simulate
    #Returns
        plotting of brownian motion using the array of distances (float) sampled from normal distribution N(0,dt)
    """
    #Set colors
       
    colors=mcolors.CSS4_COLORS
    colors=list(colors.values())

    random.shuffle(colors)

    if len(colors)<n:
        raise Exception("Too many samples, please redefine color library")
    for i in range(n):

        distances=BrownianMotionSim(time,samples=samples,seed=None)
        distances=list(distances[0,:])
        plt.plot(np.arange(0,time,time/samples),distances,color=colors[i],label="Simulation "+str(i+1))
    
    plt.xlabel("Time (s)")
    plt.ylabel("Distance")
    plt.legend(loc="best", bbox_to_anchor=(1,1))
    plt.show()

    return
```

```python
#Test functions and explore Brownian motion
time=120
plotBrownian(time,n=10,samples=100,seed=None)
```

```python
time=500
plotBrownian(time,n=5,samples=100,seed=None)
```

### Key observations

Stochastic processes are random; however, approximating underlying functions that generate the observations allows making inferences.

The further in time the inferences are made, the less reliable predictions are going to be. You can observe that in the plots based on how points diverge as the simulation continues.
